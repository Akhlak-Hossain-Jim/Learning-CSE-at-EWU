```python
# Install if needed
# !pip install missingno

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
from scipy import stats

df = pd.read_csv('Boston_Housing_assignment.csv')
print(df.head())
target_column = 'medv'
```

```python
# Samia's Columns
samia_columns = ['crim', 'zn', 'indus', 'chas']

# Subset for analysis
samia_df_subset = df[samia_columns].copy()
```

```python
# Suraiya's Columns
suraiya_columns = ['nox', 'rm','age','dis']

# Subset for analysis
suraiya_df_subset = df[suraiya_columns].copy()
```

```python
# Mehedi's Columns
mehedi_columns = ['rad', 'tax', 'ptratio','black']

# Subset for analysis
mehedi_df_subset = df[mehedi_columns].copy()
```

```python
# Akhlak's Columns
akhlak_columns = ["chas","dis","black",'lstat']

# Subset for analysis
akhlak_df_subset = df[akhlak_columns].copy()
```


## **2. Data Cleaning**

### 2.1 Missing Values (All Members)
- Identify missing values.
- Show % missing for each column.
- Apply imputation: Mean, Median, or Mode.
- **Visualizations**: Missing data barplot/heatmap.

---



```python
### DATA CLEANING SAMIA ###
print("\nMissing Data %:")
print(samia_df_subset.isnull().mean() * 100)

# Visualize missing
msno.bar(samia_df_subset)
plt.title("Missing Before Cleaning")
plt.show()

# Handle missing
for col in samia_columns:
    if samia_df_subset[col].dtype == 'object':
        samia_df_subset[col].fillna(samia_df_subset[col].mode()[0], inplace=True)
    else:
        samia_df_subset[col].fillna(samia_df_subset[col].median() if samia_df_subset[col].skew() > 1 else samia_df_subset[col].mean(), inplace=True)

# After Cleaning
msno.bar(samia_df_subset)
plt.title("Missing After Cleaning")
plt.show()
```

```python
### DATA CLEANING SURAIYA ###
print("\nMissing Data %:")
print(suraiya_df_subset.isnull().mean() * 100)

# Visualize missing
msno.bar(suraiya_df_subset)
plt.title("Missing Before Cleaning")
plt.show()

# Handle missing
for col in suraiya_columns:
    if suraiya_df_subset[col].dtype == 'object':
        suraiya_df_subset[col].fillna(suraiya_df_subset[col].mode()[0], inplace=True)
    else:
        suraiya_df_subset[col].fillna(suraiya_df_subset[col].median() if suraiya_df_subset[col].skew() > 1 else suraiya_df_subset[col].mean(), inplace=True)

# After Cleaning
msno.bar(suraiya_df_subset)
plt.title("Missing After Cleaning")
plt.show()
```

```python
### DATA CLEANING MEHEDI ###
print("\nMissing Data %:")
print(mehedi_df_subset.isnull().mean() * 100)

# Visualize missing
msno.bar(mehedi_df_subset)
plt.title("Missing Before Cleaning")
plt.show()

# Handle missing
for col in mehedi_columns:
    if mehedi_df_subset[col].dtype == 'object':
        mehedi_df_subset[col].fillna(mehedi_df_subset[col].mode()[0], inplace=True)
    else:
        mehedi_df_subset[col].fillna(mehedi_df_subset[col].median() if mehedi_df_subset[col].skew() > 1 else mehedi_df_subset[col].mean(), inplace=True)

# After Cleaning
msno.bar(mehedi_df_subset)
plt.title("Missing After Cleaning")
plt.show()
```

```python
### DATA CLEANING AKHLAK ###
print("\nMissing Data %:")
print(akhlak_df_subset.isnull().mean() * 100)
print(akhlak_df_subset.isnull().sum())

# Visualize missing
msno.bar(akhlak_df_subset)
plt.title("Missing Before Cleaning")
plt.show()

# Handle missing
for col in akhlak_columns:
    if akhlak_df_subset[col].dtype == 'object':
        akhlak_df_subset[col].fillna(akhlak_df_subset[col].mode()[0], inplace=True)
    else:
        akhlak_df_subset[col].fillna(akhlak_df_subset[col].median() if akhlak_df_subset[col].skew() > 1 else akhlak_df_subset[col].mean(), inplace=True)

# After Cleaning
msno.bar(akhlak_df_subset)
plt.title("Missing After Cleaning")
plt.show()
```

## **3. Outlier Detection**

### 3.1 Visual Detection
- Boxplots for each numerical column.

### 3.2 Statistical Detection
- Z-score or IQR method to detect outliers.
- Cap/remove outliers.
- Re-plot boxplots.

---



```python
# Detect outliers using Z-score SAMIA
samia_z = np.abs(stats.zscore(samia_df_subset))
samia_df_no_outliers = samia_df_subset[(samia_z < 3).all(axis=1)]

# Plot side-by-side boxplots
for col in samia_columns:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    sns.boxplot(x=samia_df_subset[col], ax=axes[0])
    axes[0].set_title(f'Before Outlier Removal - {col}')

    sns.boxplot(x=samia_df_no_outliers[col], ax=axes[1])
    axes[1].set_title(f'After Outlier Removal - {col}')

    plt.tight_layout()
    plt.show()
```

```python
# Detect outliers using Z-score suraiya
suraiya_z = np.abs(stats.zscore(suraiya_df_subset))
suraiya_df_no_outliers = suraiya_df_subset[(suraiya_z < 3).all(axis=1)]

# Plot side-by-side boxplots
for col in suraiya_columns:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    sns.boxplot(x=suraiya_df_subset[col], ax=axes[0])
    axes[0].set_title(f'Before Outlier Removal - {col}')

    sns.boxplot(x=suraiya_df_no_outliers[col], ax=axes[1])
    axes[1].set_title(f'After Outlier Removal - {col}')

    plt.tight_layout()
    plt.show()
```

```python
# Detect outliers using Z-score mehedi
mehedi_z = np.abs(stats.zscore(mehedi_df_subset.select_dtypes(include=[np.number]), nan_policy='omit'))
mehedi_df_no_outliers = mehedi_df_subset[(mehedi_z < 3).all(axis=1)].reset_index(drop=True)

# Plot side-by-side boxplots
for col in mehedi_columns:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    sns.boxplot(x=mehedi_df_subset[col], ax=axes[0])
    axes[0].set_title(f'Before Outlier Removal - {col}')

    sns.boxplot(x=mehedi_df_no_outliers[col], ax=axes[1])
    axes[1].set_title(f'After Outlier Removal - {col}')

    plt.tight_layout()
    plt.show()
```

```python
# Detect outliers using Z-score akhlak
akhlak_z = np.abs(stats.zscore(akhlak_df_subset.select_dtypes(include=[np.number]), nan_policy='omit'))
akhlak_df_no_outliers = akhlak_df_subset[(akhlak_z < 3).all(axis=1)].reset_index(drop=True)

# Plot side-by-side boxplots
for col in akhlak_columns:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    sns.boxplot(x=akhlak_df_subset[col], ax=axes[0])
    axes[0].set_title(f'Before Outlier Removal - {col}')

    sns.boxplot(x=akhlak_df_no_outliers[col], ax=axes[1])
    axes[1].set_title(f'After Outlier Removal - {col}')

    plt.tight_layout()
    plt.show()
```

## **4. Univariate Analysis**

### 4.1 For Each Column (All Members)
- Histogram + Boxplot.
- Normality Test (Shapiro/D’Agostino).
- Description of distribution (Normal/Skewed).

---



```python
### 3. UNIVARIATE ANALYSIS samia ###
for col in samia_columns:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(samia_df_no_outliers[col], kde=True, ax=axes[0])
    sns.boxplot(x=samia_df_no_outliers[col], ax=axes[1])
    axes[0].set_title(f'Histogram - {col}')
    axes[1].set_title(f'Boxplot - {col}')
    plt.show()

    stat, p = stats.shapiro(samia_df_no_outliers[col])
    print(f"{col} - Shapiro-Wilk Test p-value: {p:.4f}")
    print("Probably Normal" if p > 0.05 else "Probably Not Normal")
```

```python
### 3. UNIVARIATE ANALYSIS suraiya ###
for col in suraiya_columns:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(suraiya_df_no_outliers[col], kde=True, ax=axes[0])
    sns.boxplot(x=suraiya_df_no_outliers[col], ax=axes[1])
    axes[0].set_title(f'Histogram - {col}')
    axes[1].set_title(f'Boxplot - {col}')
    plt.show()

    stat, p = stats.shapiro(suraiya_df_no_outliers[col])
    print(f"{col} - Shapiro-Wilk Test p-value: {p:.4f}")
    print("Probably Normal" if p > 0.05 else "Probably Not Normal")
```

```python
### 3. UNIVARIATE ANALYSIS mehedi ###
for col in mehedi_columns:
    # Convert the column to numeric, handling errors
    mehedi_df_no_outliers[col] = pd.to_numeric(mehedi_df_no_outliers[col], errors='coerce')

    # Drop rows with NaN values after conversion
    mehedi_df_no_outliers.dropna(subset=[col], inplace=True)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(mehedi_df_no_outliers[col], kde=True, ax=axes[0])
    sns.boxplot(x=mehedi_df_no_outliers[col], ax=axes[1])
    axes[0].set_title(f'Histogram - {col}')
    axes[1].set_title(f'Boxplot - {col}')
    plt.show()

    stat, p = stats.shapiro(mehedi_df_no_outliers[col])
    print(f"{col} - Shapiro-Wilk Test p-value: {p:.4f}")
    print("Probably Normal" if p > 0.05 else "Probably Not Normal")
```

```python
### 3. UNIVARIATE ANALYSIS akhlak ###
for col in akhlak_columns:
    # Convert the column to numeric if it's not already, handling errors
    akhlak_df_no_outliers[col] = pd.to_numeric(akhlak_df_no_outliers[col], errors='coerce')

    # Drop rows with NaN values after conversion, which might have resulted from failed numeric conversion
    akhlak_df_no_outliers.dropna(subset=[col], inplace=True)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(akhlak_df_no_outliers[col], kde=True, ax=axes[0])
    sns.boxplot(x=akhlak_df_no_outliers[col], ax=axes[1])
    axes[0].set_title(f'Histogram - {col}')
    axes[1].set_title(f'Boxplot - {col}')
    plt.show()

    stat, p = stats.shapiro(akhlak_df_no_outliers[col])
    print(f"{col} - Shapiro-Wilk Test p-value: {p:.4f}")
    print("Probably Normal" if p > 0.05 else "Probably Not Normal")
```
